```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

You need to implement a TF-IDF vectorizer to convert a collection of docs
into TF-IDF vectors. You can use the sklearnâ€™s inbuild dataset
fetch_20newsgroups.

Create a function to calculate the cosine similarity between two TF-IDF vectors.

Implement a document similarity search function that takes a document as input
and returns a list of docs ranked by their similarity to the input document.

```python
def tfidf_vectorizer(docs):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(docs)
    return tfidf_matrix, vectorizer

def cosine_similarity_matrix(tfidf_matrix):
    return cosine_similarity(tfidf_matrix)

def doc_sim_search(q_doc, tfidf_matrix, vectorizer, docs, num_results=5):
    query_vector = vectorizer.transform([q_doc])
    similarities = cosine_similarity(query_vector, tfidf_matrix)
    
    # Get indices of top similar docs
    similar_indices = similarities.argsort(axis=1)[:, -num_results:][0]
    
    # Return the most similar docs
    similar_docs = [docs[i] for i in similar_indices]
    
    return similar_docs

# Example usage with fetch_20newsgroups dataset
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
docs = newsgroups.data

# Vectorize docs
tfidf_matrix, vectorizer = tfidf_vectorizer(docs)

# Example query document
q_doc = "Students scramble as Codeup closure leaves them in debt, unsure how to complete their studies"
similar_docs = doc_sim_search(q_doc, tfidf_matrix, vectorizer, docs)

print(f"Query Document: {q_doc}")
print("Similar Documents:")
for i, doc in enumerate(similar_docs, start=1):
    print(f"{i}. {doc}")
```
